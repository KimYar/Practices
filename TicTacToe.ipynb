```
import random
board = [" " for i in range(9)]
board_list = [0,1,2,3,4,5,6,7,8]


def choosing_mark():
     global choice
     global opponent
     global player
     opponent = " "
     player = " "
     choice = input("What would you like? X or O? ").strip().upper()
     while choice != "X" and choice != "O":
       choice = input("What would you like? X or O? ").strip().upper()
     game_board()
     if choice == "X":
       opponent = "O"
       player = "X"  
     else:
       opponent = "X"
       player = "O"

def game_board():

  row1 = "{} | {} | {}".format (board[0], board[1],board[2])
  row2 = "{} | {} | {}".format (board[3], board[4],board[5])
  row3 = "{} | {} | {}".format (board[6], board[7],board[8])
  print()
  print(row1)
  print(row2)
  print(row3)
  print()

def getting_number_player():
  global position
  entered_value = input("Please enter your position's number (0-8): ").strip()
  while not entered_value.isnumeric():
    entered_value = input("Please enter your position's number (0-8): ").strip()
  position = int(entered_value)
  while position not in board_list:
    position = int(input("Please enter your position's number (0-8): ").strip())
  board[position] = player
  board_list.remove(position)
  game_board()
  
    
def getting_number_computer():
  random_num = random.choice(board_list)
  board[random_num] = opponent
  board_list.remove(random_num)
  game_board()

def is_victory():
  if is_line_same_shape(0, 1, 2) or \
     is_line_same_shape(3, 4, 5) or \
     is_line_same_shape(6, 7, 8) or \
     is_line_same_shape(0, 3, 6) or \
     is_line_same_shape(1, 4, 7) or \
     is_line_same_shape(2, 5, 8) or \
     is_line_same_shape(0, 4, 8) or \
     is_line_same_shape(2, 4, 6):
     return True
  else:
    return False

def is_line_same_shape(i, j, k):
  return (board[i] == board[j] and board[j] == board[k] and board[i] != " ")

def is_move():
  if len(board_list) != 0:
    return True
  else: 
    return False
        
def game_condition():
  if position != 4:
    board[4] = opponent
    game_board()  
    board_list.remove(4)  
    while is_move():  
      getting_number_player()
      if is_victory():
        print("Congratulation! You are the winner!")
        break
      else:
        if is_move():
          getting_number_computer()
          if is_victory():
            print("Sorry! You lost!")
            break
        else:
          print("Game Over!!!")
    
  else:    
    getting_number_computer()
    while True:  
      getting_number_player()
      if is_victory():
        print("Congratulation! You are the winner!")
        break
      else:
        if is_move():
          getting_number_computer()
          if is_victory():
            print("Sorry! You lost!")
            break
        else:
          print("Game Over!!!")


choosing_mark()

if player == "X":
  getting_number_player()
  game_condition()
  
else:
  getting_number_player()
  game_condition()
```
